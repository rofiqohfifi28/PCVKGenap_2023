```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import math
from google.colab.patches import cv2_imshow
from PIL import Image as im
```

```python
img = cv.imread('/content/drive/MyDrive/PCVK/Images/jm.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
def convolution2d(image, kernel, stride, padding):
  image = np.pad(image, padding, mode='constant', constant_values=0)

  kernel_height, kernel_width = kernel.shape
  padded_height, padded_width = image.shape

  output_height = (padded_height - kernel_height) // stride + 1
  output_width = (padded_width - kernel_width) // stride + 1

  new_image = np.zeros((output_height, output_width)).astype(np.float32)

  # convolution process
  for y in range(0, output_height):
    for x in range(0, output_width):
      new_image[y][x] = np.sum(image[y * stride:y * stride + kernel_height, x * stride:x * stride + kernel_width] * kernel).astype(np.float32)
  return new_image
```

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

FILTERING MEAN

```python
kernal_mean = np.array([[1/5,1/5,1/5],
                       [1/5,1/5,1/5],
                       [1/5,1/5,1/5]])
```

```python
print('Gambar Asli')
cv2_imshow(img_gray)
print('Hasil Mean')
cv2_imshow(convolution2d(img_gray,kernal_mean,1,2))
```

FILTERING MEDIAN

```python
kernal_median = np.array([[1/5,1/7,1/5],
                       [1/5,1/7,1/5],
                       [1/5,1/7,1/5]])
```

```python
print('Gambar Asli')
cv2_imshow(img_gray)
print(' Hasil Median')
cv2_imshow(convolution2d(img_gray,kernal_median,1,2))
```

FILTERING MODUS

```python
kernal_modus = np.array([[1/5,1/7,1/5],
                       [1/7,1/4,1/7],
                       [1/7,1/7,1/1]])

```

```python
print('Gambar Asli')
cv2_imshow(img_gray)
print('Hasil Modus')
cv2_imshow(convolution2d(img_gray,kernal_modus,1,2))
```

HIGH PASS

```python
kernal_high = np.array([[0,-1,0],
                       [-1,5,-1],
                       [0,-1,0]])
```

```python
print('Gambar Asli')
cv2_imshow(img_gray)
print('Hasil High Pass')
cv2_imshow(convolution2d(img_gray,kernal_high,1,2))
```
